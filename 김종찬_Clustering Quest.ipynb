```python
altered_dataset = dataset.drop(columns = ['Country', 'Happiness.Rank'])
```

```python
scaler1 = preprocessing.MinMaxScaler()
scaled_dataset1 = scaler1.fit_transform(altered_dataset)
scaled_dataset1 = pd.DataFrame(scaled_dataset1, columns = ['Happiness.Score', 'Whisker.high',
       'Whisker.low', 'Economy..GDP.per.Capita.', 'Family',
       'Health..Life.Expectancy.', 'Freedom', 'Generosity',
       'Trust..Government.Corruption.', 'Dystopia.Residual'])
```

```python
scaler2 = preprocessing.StandardScaler()
scaled_dataset2 = scaler2.fit_transform(altered_dataset)
scaled_dataset2 = pd.DataFrame(scaled_dataset2, columns =  ['Happiness.Score', 'Whisker.high',
       'Whisker.low', 'Economy..GDP.per.Capita.', 'Family',
       'Health..Life.Expectancy.', 'Freedom', 'Generosity',
       'Trust..Government.Corruption.', 'Dystopia.Residual'])
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 10))
ax1.set_title('Before Scaling')
sns.kdeplot(altered_dataset['Happiness.Score'], ax=ax1)
sns.kdeplot(altered_dataset['Whisker.high'], ax=ax1)

ax2.set_title('After Min-Max Scaling')
sns.kdeplot(scaled_dataset1['Happiness.Score'], ax=ax2)
sns.kdeplot(scaled_dataset1['Whisker.high'], ax=ax2)
plt.show()
```

```python
testdata1=altered_dataset.loc[:,['Trust..Government.Corruption.','Happiness.Score']].values
testdata2=scaled_dataset1.loc[:,['Trust..Government.Corruption.','Happiness.Score']].values
testdata3=scaled_dataset2.loc[:,['Trust..Government.Corruption.','Happiness.Score']].values
```

```python
from sklearn.cluster import KMeans
score1=[]
```

```python
for i in range(1,11):
    kmeans1 = KMeans(n_clusters = i, init = 'k-means++', random_state =0)
    kmeans1.fit(testdata1)
    score1.append(kmeans1.inertia_)
```

```python
plt.figure(figsize=(8,8))
plt.plot(range(1,11),score1)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeansmodel1 = KMeans(n_clusters=3,init='k-means++', random_state=0)
y_kmeans1 = kmeansmodel1.fit_predict(testdata1)
```

```python
plt.figure(figsize=(12, 12))
plt.scatter(testdata1[y_kmeans1 == 0, 0], testdata1[y_kmeans1 == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(testdata1[y_kmeans1 == 1, 0], testdata1[y_kmeans1 == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(testdata1[y_kmeans1 == 2, 0], testdata1[y_kmeans1 == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.title('Clusters of customers')
plt.xlabel('Trust..Government.Corruption.')
plt.ylabel('Happiness.Score')
plt.legend()
plt.show()
```
