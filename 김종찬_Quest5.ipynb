```python
y = (y==0).astype(np.int)
y = y.reshape(-1,1)
```

```python
log_reg = LogisticRegression()
log_reg.fit(X_meanradius,y)
```

```python
X_new = np.linspace(5, 25, 1000).reshape(-1,1)
y_prob = log_reg.predict_proba(X_new)
y_prob
```

```python
plt.plot(X_new, y_prob[:,1], 'g-', label='breast_canser')
plt.plot(X_new, y_prob[:,0], 'b-', label='Not breast_canser')
```
